## Task 1 :- Merge The 12 Months of Sales Data Into A Single CSV File

```python
files = [ file for file in os.listdir('./data/')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./data/" + file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data
```

```python
all_months_data.tail()
```

```python
# Export to a single CSV

all_months_data.to_csv('all_data.csv', index = False)
```

```python
# Check
all_data = pd.read_csv('./all_data.csv')
all_data
```

## Clean Data

```python
all_data.isna().sum()
```

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df
```

```python
all_data = all_data.dropna(how='all')
all_data
```

```python
# No NAN values
all_data.isna().sum()
```

### Convert orderdate string type to integer

```python
# convert orderdate string type to integer

all_data['Month'] = all_data['Order Date'].str[0:2]
all_data
```

```python
all_data["Month"] = all_data["Month"].astype('int32')
```

Getting this error of having values of OR, Check if there are any OR values in OrderDate Columns

```python
temp_df = all_data[all_data['Order Date'].str[0:2]=='Or']
temp_df
```

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
all_data
```

### Augment Data With Additional Columns

#### Task2 :-  Add Month Column

```python
all_data["Month"] = all_data["Month"].astype('int32')
all_data
```

#### Add A Sales Column

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
all_data
```

```python
all_data["Sales"] = all_data["Quantity Ordered"] * all_data["Price Each"]
all_data
```

### Question 1 :- What was the best month for sales ? How much was earned that month?

```python
sales_sum = all_data.groupby('Month').sum('Sales')
sales_sum
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, sales_sum['Sales'])
plt.xlabel('Month')
plt.ylabel(' Sales in USD')
plt.show()
```

### Question 2 :-   Which city has highest sales 

```python
all_data.head()
```

#### Convert Purchase address into City and State for better analysis

```python
all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])
all_data

```

```python
# state

all_data['State']   = all_data['Purchase Address'].apply(lambda x: x.split(',')[2]).str[0:3]
all_data['Pincode'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[2]).str[4:9]
all_data

```

```python
all_data.groupby(['City']).sum('Sales')
```

```python
# plot a bar graph

results = all_data.groupby(['City']).sum('Sales')

cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(rotation='vertical')
plt.show()
```

### Question 3 :-  What time should we display advertisements to maximize liklihood of customers buying product?

```python
all_data.head()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Min']  = all_data['Order Date'].dt.minute
all_data
```

```python
# Plot A Graph

hours  = [ hour for hour, df in all_data.groupby('Hour')]
order_count = all_data.groupby(['Hour']).count()

plt.plot(hours, order_count)
plt.xlabel('Hours')
plt.ylabel('Order Count')
plt.xticks(hours)
plt.grid()
plt.show()
```

### Question 4:- What Products are most often sold together?

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df =df[['Order ID','Grouped']].drop_duplicates()

df.head(20)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

count.most_common(10)
```
